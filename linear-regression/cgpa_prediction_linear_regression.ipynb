```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df= pd.read_csv('hour_cgpa_dataset.csv')
df
```

```python
X = df.drop('cgpa', axis=1)
y = df['cgpa']
X
```

```python
y
```

```python
plt.Figure(figsize=(4,3))
plt.scatter(X, y)
```

```python
#step 1: declare an object of the model
model = LinearRegression()
```

```python
#step 2: fit the model to the data
model.fit(X,y)
```

```python
#step 3:predict the target variable using the model
y_pred = model.predict(X)
y_pred
```

```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2s = r2_score(y, y_pred)
```

```python
print(f'mean squared error: {mse}, mean absolute error: {mae}, R2 score: {r2s}')


```

```python
m = model.coef_
b = model.intercept_
print(f'model coefficient: {m}, intercept: {b}')
```

```python
X1 = []
y1 = []
for i in range(80, 170, 5):   
    X1.append(i)
    y1.append(m[0]*i + b)
plt.Figure(figsize=(4,3))

plt.scatter(X, y)

# Plot the line (y = m * x + b)
plt.plot(X1, y1, color='red', label="Fitted Line")
plt.xlabel("Hours Studied (X)")
plt.ylabel("CGPA (y)")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
input_hours = 120
predicted_cgpa = model.predict([[input_hours]])
print(f'predicted CGPA for {input_hours} hours of study: {predicted_cgpa[0]}')
```
