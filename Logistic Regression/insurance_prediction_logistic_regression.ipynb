```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv('insurance_data.csv')
print(df.head(7))
```

```python
#input feture
X = df.drop(['bought', 'bought y/n'], axis=1)

#target feature
y = df['bought y/n']
X
```

```python
#Plot the data  
plt.figure(figsize=(10, 6))
plt.scatter(X,y)
plt.xlabel('Age')
plt.ylabel('Bought Insurance')  
```

```python
lr = LogisticRegression()
lr.fit(X, y)
y_pred = lr.predict(X)
y_pred
```

```python
cm = confusion_matrix(y, y_pred)
print(cm)   
cr = classification_report(y, y_pred)
print(cr)   
```

```python
input_age = int(input("Enter the age: "))
input_data = np.array([[input_age]])    
out=lr.predict(input_data)
if out[0] == 1:
    print("The person is likely to buy insurance.") 
else:
    print("The person is not likely to buy insurance.") 
```
